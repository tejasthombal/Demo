```python
import numpy as np
```

```python
size = int(input("Enter the size:"))
x = []
y = []
print("Enter the values of X:")
for i in range(size):
  k = int(input())
  x.append(k)

print("Enter the values of Y:")
for j in range(size):
  s = int(input())
  y.append(s)

print(x)
print(y)
```

```python
x=np.array(x)
y=np.array(y)
```

```python
# Practical 7
def euclidean_distance(x,y):
  return sum((x-y)**2)**(1/2)
```

```python
print("Euclidean Distance :",euclidean_distance(x,y))
```

```python
def manhattan_distance(x,y):
  return sum(abs(x-y))
```

```python
print("Manhattan Distance:",manhattan_distance(x,y))
```

```python
# Practical 8
def minkowski(x,y,p):
  return sum(abs(x-y)**p)**(1/p)
```

```python
print('Minkowski Distance:',minkowski(x,y,3))
```

```python
a = np.dot(x,y)
```

```python
x_norm = np.linalg.norm(x)
y_norm = np.linalg.norm(y)
```

```python
cosine = a /(x_norm * y_norm)
```

```python
print('Cosine:',cosine)
```
